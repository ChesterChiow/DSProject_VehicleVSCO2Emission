### About Dataset
**Make**: Company of the vehicle  
**Model**: Car Model  
**Vehicle Class**: Class of vehicle depending on their utility, capacity and weight

**Cylinders**: Number of cylinders (engineâ€™s power unit where fuel combustion takes place to generate power that moves a car.)  
**Transmission**: Transmission type with number of gears (3-10) 
- A = Automatic
- AM = Automated manual
- AS = Automatic with select shift
- AV = Continuously variable   
- M = Manual

**Fuel Type**: Type of fuel used
- X = Regular gasoline
- Z = Premium gasoline
- D = Diesel
- E = Ethanol (E85)
- N = Natural gas

**Fuel Consumption City (L/100 km)**: Fuel consumption in city roads in litres per 100 kilometres  
**Fuel Consumption Hwy (L/100 km)**: Fuel consumption in highways in litres per 100 kilometres  
**Fuel Consumption Comb (L/100 km)**: The combined fuel consumption (55% city, 45% highway) in litres per 100 kilometres  
**Fuel Consumption Comb (mpg)**: The combined fuel consumption (55% city, 45% highway) in mile per gallon  
**CO2 Emissions(g/km)**: The tailpipe emissions of carbon dioxide in grams per kilometre for combined city and highway driving

# Data Cleaning

```python
# Create a copy of the dataset
df_clean = df.copy()
```

### Check for Missing Data

```python
df.isnull().sum()
```

There are no missing data in any of the columns.

### Check for Duplicated Data

```python
print("Number of duplicated data:", df.duplicated().sum())
```

```python
# Extract and Display the rows that are duplicated
dfDuplicated = df[df.duplicated() == True]
dfDuplicated.head()
```

```python
# Drop the duplicated rows
df_clean = df.drop_duplicates()
print("Data dims : ", df_clean.shape)
```

### Create a new column for Number of Gears

```python
# Create a copy of the dataset
df_clean2 = df_clean.copy()

# Extract number of gears from 'Transmission' column
df_clean2['Number of Gears'] = df_clean2['Transmission'].str.extract(r'(\d+)', expand=False)

# Remove number of gears from 'Transmission' column
df_clean2['Transmission'] = df_clean2['Transmission'].str.replace(r'\d+', '', regex=True)

df_clean2.head()
```

```python
# Check for missing values in number of gears column
print("Number of rows with missing values:",df_clean2["Number of Gears"].isnull().sum())
```

```python
# Extract rows with null values in the "Number of Gears" column
rows_with_null_gears = df_clean2[df_clean2["Number of Gears"].isnull()]

# Display the extracted rows
rows_with_null_gears.head()
```

The percentage of missing values in number of gears column is 3.84% (241/6282) which is small and may not signficiantly impact out analysis. Hence, we have chosen to drop these rows to simplify data analysis and processing. 

```python
# Drop the rows with null values in the "Number of Gears" column
df_clean2.dropna(subset=["Number of Gears"], inplace=True)
print("Data dims : ", df_clean2.shape)
```

### Drop Unnecessary Columns
Fuel Consumption Comb (L/100 km) and Fuel Consumption Comb (mpg) measures the same quantity â€” fuel consumption - but in different units.
We will use Fuel Consumption Comb (L/100 km) only for our investigation.

```python
# Create a copy of the dataset
df_clean3 = df_clean2.copy()

# Drop 'Fuel Consumption Comb (mpg)' column
df_clean3.drop(columns=['Fuel Consumption Comb (mpg)'], inplace=True)
df_clean3.head()
```
